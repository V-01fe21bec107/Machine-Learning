```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([10,48,23,56,89,9,32,45])
plt.scatter(x,y,marker='*',c='b')
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=np.array([10,20,30,40,50,60,70,80]).reshape((-1,1))
y=np.array([1,14,23,36,49,56,62,75])
plt.scatter(x,y,marker='*',c='b')
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='red',marker='*')
print("mse:",mse)
print("intercept:",SLR.intercept_)
print("slope:",SLR.coef_)
```


A. Instead of directly giving values, randomly generate x and y using random module and do linear regression.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x=np.random.randint(1,100,10).reshape((-1,1))
y=np.random.randint(1,200,10)
plt.scatter(x,y,marker='*',c='b')
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='red',marker='*')
print("mse:",mse)
print("intercept:",SLR.intercept_)
print("slope:",SLR.coef_)
```
